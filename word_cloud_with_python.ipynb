```python
def show_wordcloud(title,filepath):
  title=wikipedia.search(title)[0]
  page=wikipedia.page(title)
  content=page.content
  background=np.array(Image.open(filepath))
  wc=WordCloud(background_color='white',mask=background,max_words=200,stopwords=storwo,colormap='rainbow',contour_width=0.5).generate(content)
  plt.figure(figsize=(18,12))
  plt.axis('off')
  plt.imshow(wc)
  plt.savefig('IISCfig.png')
```

```python
show_wordcloud('IISc','/content/IISc_Master_Logo_White.jpg')
```

```python
def show_wordcloud(title,shapefile,filesave_name):
  content=title
  background=np.array(Image.open(shapefile))
  wc=WordCloud(background_color='white',mask=background,colormap='ocean',repeat=True,random_state=42).generate(content)##,contour_color='black',contour_width=0.05
  plt.figure(figsize=(18,12))
  plt.axis('off')
  plt.imshow(wc)
  plt.savefig(filesave_name)
```
